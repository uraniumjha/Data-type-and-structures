8.What is a hash table, and how does it relate to dictionaries in Python ?
 - A hash table is a data structure that stores key-value pairs. It uses a hash function to compute an index (or hash code) for each key mapping it to a specific bucket or slot in memory. This allows for efficient operations like insertion deletion and lookup with an average time complexity of O(1)

10.Why strings are immutable in python
  - Strings are immutable in Python to ensure thread safety, memory efficiency, and reliable behavior when used as hashable keys in data structures like dictionaries and sets

11.What advantages do dictionaries offer over lists for certain tasks ?
  - Dictionaries offer faster lookups and efficient key-value pairing making them ideal for tasks that require quick access by keys or mapping relationships. Unlike lists, dictionaries provide O(1) average time complexity for lookups whereas lists require O(n) for searching elements

12.Describe a scenario where using a tuple would be preferable over a list?
   - A tuple would be preferable over a list in a scenario where you need to store a fixed collection of related values such as the coordinates of a location (latitude, longitude) and you want to ensure the data remains unchanged. Since tuples are immutable, they guarantee data integrity and can also be used as keys in dictionaries unlike lists.
